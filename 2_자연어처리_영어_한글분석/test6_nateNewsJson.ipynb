```python
# 주의사항, 문자열 안에 특수문자 부분 체크 잘하기.
# create table newsTable (
# id int not null auto_increment,
# title varchar(200),
# publisher varchar(200),
# newsDate varchar(10),
# newsTime varchar(10),
# newsDetail varchar(200),
# newsImgUrl varchar(200),
# primary key(id)
# );
# regex 이용해서, re.sub(매개변수1,매개변수2,매개변수3)
# 매개변수1 : 특정 조건
# 매개변수2 : 대체할 문자
# 매개변수3 : regex를 적용할 문장.
```

```python
import time
import bs4
import urllib.request
# import pymysql
import re
import json
def clean_text(inputString):
    text_rmv = re.sub(
        '[-=+,#/\?:^.@*\"※~ㆍ!』‘|\(\)\[\]`\'…》\”\“\’·]', ' ', inputString)
    return text_rmv

# 전역변수, 기사에서 가지고 올 정보만 선택한 변수.
imgLinkUrl, subject, link, press, pDate, pTime = "", "", "", "", "", ""

page = 1
count = 1
max_count = 1000
nateUrl = "https://news.nate.com/recent?mid=n0100&type=c&date=20240813&page="
# htmlObject = urllib.request.urlopen(nateUrl)
# webPage = htmlObject.read()
# bsObject = bs4.BeautifulSoup(webPage, 'html.parser')

#각 기사 딕션너리 객체, 
# 각 기사를 리스트에 담기. 
news_list = []
while count <= max_count:
    try:
        newsUrl = nateUrl + str(page)
        print(f"newsUrl: {newsUrl}")
        htmlObject = urllib.request.urlopen(newsUrl)
        webPage = htmlObject.read()
        bsObject = bs4.BeautifulSoup(webPage, 'html.parser')
        
        tag_list = bsObject.findAll('div', {'class': 'mlt01'})
        print('###### 실시간 뉴스 속보 #######')
        for tag in tag_list:
            if count > max_count:
                break
            
            subject = tag.find('h2', {'class': 'tit'}).text
            subject = clean_text(subject)
            print(f"subject: {subject}")
            print('================================')
          
            news_item = {'subject': subject}
            news_list.append(news_item)
            count += 1
        
        page += 1
        time.sleep(1)  # 웹사이트 요청 부담을 줄이기 위해 1초 대기
        print(f"수집된 뉴스 수: {count-1}, 현재 페이지: {page}")
        
    except Exception as e:
        print(f"오류 발생: {e}")
        break

# JSON 파일로 저장
with open('sampleNateNews.json', 'w', encoding='utf-8') as json_file:
    json.dump(news_list, json_file, ensure_ascii=False, indent=4)

print("완료! JSON 파일이 생성되었습니다.")
```
